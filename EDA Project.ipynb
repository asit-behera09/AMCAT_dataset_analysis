```python
# Average age of the employees:
from datetime import date

def calculate_age(birth_date):
    today = date.today()
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return age

# Example usage:
print('Average age of employee is :',pd.to_datetime(df['DOB']).apply(calculate_age).mean())
```

```python
def one_way_anova(all_groups, significance_level):
    stat, p_val = stats.f_oneway(*all_groups)
    
    print('stat=%.3f, p_val=%.3f' % (stat, p_val))
    
    if p_val < significance_level:
        print('Reject null hypothesis (H0): There is a significant difference between the groups.')
    else:
        print('Fail to Reject null hypothesis (H0): There is no significant difference between the groups.')
```

```python
DegreeBtech = df[df['Degree']=='B.Tech/B.E.']['Salary'].sample(n=100,replace=True)
DegreeMSc = df[df['Degree']=='M.Sc. (Tech.)']['Salary'].sample(n=100,replace=True)
DegreeMTech = df[df['Degree']=='M.Tech./M.E.']['Salary'].sample(n=100,replace=True)
DegreeMCA = df[df['Degree']=='MCA']['Salary'].sample(n=100,replace=True)

all_groups = [DegreeBtech, DegreeMSc, DegreeMTech,DegreeMCA]
one_way_anova(all_groups, significance_level=0.05)
```

```python
def func(x):
    if x>=40 and x<60:
        return 'D'
    elif x>=60 and x<75:
        return 'C'
    elif x>=75 and x<85:
        return 'B'
    elif x>=85:
        return 'A'
df['12Grade'] = df['12percentage'].apply(func=func)
```

```python
pd.crosstab(df['12Grade'],df['CollegeTier']).plot(kind='bar')
plt.ylabel('Frequency')
plt.show()
```

```python
def chi_square(freq_table, significance_level):
    stat, p_val, dof, expected = stats.chi2_contingency(freq_table)
    
    print('stat=%.3f, p=%.3f' % (stat, p_val))
    
    if p_val < significance_level:
        print('Reject null hypothesis (H0): Probably dependent')
    else:
        print('Fail to Reject null hypothesis (H0): Probably independent')
```

```python
freq_table = pd.crosstab(df['12Grade'],df['CollegeTier'])
freq_table
```

```python
chi_square(freq_table, significance_level=0.05)
```
